```python
import numpy as np
%matplotlib inline
from google.colab.patches import cv2_imshow
import cv2
```

```python
image=cv2.imread('/content/people.jpg')
```

```python
print(cv2_imshow(image))
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
print(cv2_imshow(gray))
```

```python
haar_cascade=cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
faces=haar_cascade.detectMultiScale(gray,1.1,4)
```

```python
for (x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
  cv2_imshow(image)
  cv2.waitKey(0)
```
